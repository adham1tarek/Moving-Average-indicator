```python
import pandas as pd
import datetime 
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
print('Enter the company:')
Ticker= input()
Ticker_df = yf.download(Ticker, 
                      start='2021-01-01', 
                      end='2021-12-30', 
                      progress=False,
)
print('Enter the days number:')
N= input()
N = int(N)
sum=0
for num in range(0, N+1, 1):
    sum = sum+Ticker_df.Close.array[num]
average = sum / N
print(average)
```
